```python
import pandas as pd
base = pd.read_csv('../dataset/dataset_train.csv')
```

```python
base
```

```python
day = []
month = []
year = []
for date in base.negotiation_date.values:
    date = date.split('/')
    day.append(date[0])
    month.append(date[1])
    year.append(date[2])
```

```python
base['day'] = day
base['month'] = month
base['year'] = year
```

```python
base['negotiation_date'] = pd.to_datetime(base['negotiation_date'], yearfirst=True).dt.strftime("%Y-%d-%m")
```

```python
base['product'].unique()
```

```python
for year in base.year.unique():
    year_all = base[base['year']==year]
    print('YEAR ', year)
    for month in year_all.month.unique():
        print('MONTH', month)
        print(year_all[year_all['month']==month].groupby('day').sum())
        print('NUM OF DAYS: ', len(year_all[year_all['month']==month].groupby('day').sum()))
```

```python
for year in base.year.unique():
    year_all = base[base['year']==year]
    for month in year_all.month.unique():
        month_all = year_all[year_all['month']==month]
        days = month_all.day.values
        for day in range(1, 31):
            if day not in days:
                new_row = {'negotiation_date': year + '-' + str(month).zfill(2) + '-' + str(day).zfill(2), 'sold_price': 0.0, 'product': None, 'day': day, 'month': month, 'year': year}
                base = base.append(new_row, ignore_index=True)
```
