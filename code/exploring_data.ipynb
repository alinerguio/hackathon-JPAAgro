```python
import pandas as pd
base = pd.read_csv('../dataset/dataset_train.csv')
```

```python
base
```

```python
day = []
month = []
year = []
for date in base.negotiation_date.values:
    date = date.split('/')
    day.append(date[0])
    month.append(date[1])
    year.append(date[2])
```

```python
base['day'] = day
base['month'] = month
base['year'] = year
```
